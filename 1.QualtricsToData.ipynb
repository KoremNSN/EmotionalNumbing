```python
def extract_en(file_ext):
    df = pd.read_csv('../Data/Raw/participants'+str(file_ext)+'.csv', skiprows=1)
    
    # make sure everyone has critria A
    try:
        df = df[df['trauma_exposure'] != 'no_trauma']
    except:
        print('noA in the file')
        
    # new names for questionnaires items 
    dem_names = ['ID', 'Age', 'Sex', 'Ethnicity','race','education','marital']
    pcl = []
    asi = []
    phq = []
    gad = []

    for i in range(20):
        pcl.append('pcl'+str(i+1))
        if i<16:
            asi.append('asi'+str(i+1))
            if i<9:
                phq.append('phq'+str(i+1))
                if i<7:
                    gad.append('gad'+str(i+1))
    
    # zip new names and old names for column name change
    demD = dict(zip(demo, dem_names))
    pclD = dict(zip(PCL, pcl))
    asiD = dict(zip(ASI, asi))
    phqD = dict(zip(PHQ, phq))
    gadD = dict(zip(GAD, gad))
    fields = {**demD,**pclD,**asiD,**phqD,**gadD}
    
    # columns to keep
    col = demo+PCL+PICS+ASI+PHQ+GAD
    df = df[col]
    
    # rename columns for shorter names
    df = df.rename(columns=fields)
    
    NewCols = []
    for i in range(df.shape[1]):
        if (i > 26) & (i < 67):
            NewCols.append(df.columns[i].rstrip(' - Please rate how pleasant or unpleasant you find the image:'))
        else:
            NewCols.append(df.columns[i])

    # rename image columns
    df.columns = NewCols
        
    # remove participants with missing values in PCL or images
    df = df.dropna(subset=['pcl1', 'People_226_'])
    df = df.reset_index(drop=True)
    
    
    # get 8 cluster system of PTSD + probable diagnosis     
    df['ReexpInternal'] = df['pcl1'] + df['pcl2'] + df['pcl3']
    df['ReexpExternal'] = df['pcl4'] + df['pcl5']
    df['Avoidance'] = df['pcl6'] + df['pcl7']
    df['NegativeAffect'] = df['pcl8'] + df['pcl9'] + df['pcl10'] + df['pcl11']
    df['EmotionalNumbing'] = df['pcl12'] + df['pcl13'] + df['pcl14']
    df['Externalizing'] = df['pcl15'] + df['pcl16']
    df['AnxArousal'] = df['pcl17'] + df['pcl18']
    df['DysArousal'] = df['pcl19'] + df['pcl20']
    df['PCL'] = (df['ReexpInternal'] + df['ReexpExternal'] + 
                 df['Avoidance'] + df['NegativeAffect'] + 
                 df['EmotionalNumbing'] + df['Externalizing'] + 
                 df['AnxArousal'] + df['DysArousal'])
    
    # add PCL>=33
    df['PTSD'] = [((df['PCL']>=33)) & 
                  ((df['pcl1']>=2) | (df['pcl2']>=2) | (df['pcl3']>=2) | (df['pcl4']>=2) | (df['pcl5']>=2)) & 
                  ((df['pcl6']>=2) | (df['pcl7']>=2)) &
                  (((df['pcl8']>=2).astype(int) + (df['pcl9']>=2).astype(int) + (df['pcl10']>=2).astype(int) + (df['pcl11']>=2).astype(int) + (df['pcl12']>=2).astype(int) + (df['pcl13']>=2).astype(int) + (df['pcl14']>=2).astype(int))>=2) &
                   (((df['pcl15']>=2).astype(int) + (df['pcl16']>=2).astype(int) + (df['pcl17']>=2).astype(int) + (df['pcl18']>=2).astype(int) + (df['pcl19']>=2).astype(int) + (df['pcl20']>=2).astype(int))>=2)][0]

    df = df.drop(columns=pcl)
    
    # get probable diagnosis of MDD
    df['MDD'] = 0
    for i, row in df.iterrows():
        if (int(df['phq1'][i]>=2) + int(df['phq2'][i]>=2)) > 0:
            if (int(df['phq3'][i]>=2) + int(df['phq4'][i]>=2) + 
                int(df['phq5'][i]>=2) + int(df['phq6'][i]>=2) +
                int(df['phq7'][i]>=2) + int(df['phq8'][i]>=2) +
                int(df['phq9'][i]>=1)) >=5:
                df.at[i,'MDD'] = 1
    df['PHQ'] = df[phq].sum(axis=1)
    df = df.drop(columns=phq)
    
    # get anxiety score
    df['ASI'] = df[asi].sum(axis=1)
    df['GAD'] = df[gad].sum(axis=1)
    
    df = df.drop(columns=asi)
    df = df.drop(columns=gad)
    
    
    #export to csv
    df.to_csv('../Data/extracted/Cohort'+str(file_ext)+'.csv', index=False)
```
